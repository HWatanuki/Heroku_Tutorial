1) Loading the Iris dataset as a pandas dataframe

```python
import numpy as np
import pandas as pd
iris = pd.read_csv("./sample_data/Iris.csv")
iris.head()
```

2) Define x and y for training purposes

```python
X = iris.drop('Species',axis=1)
y = iris['Species']
print(X)
print(y)
```

3) Encoding the species

```python
from sklearn.preprocessing import LabelBinarizer
encoder = LabelBinarizer()
y = encoder.fit_transform(y)
print(y)
```

4) Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_X = scaler.fit_transform(X)
print(scaled_X)
```
